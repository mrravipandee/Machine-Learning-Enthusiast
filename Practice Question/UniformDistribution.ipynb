 #### A factory produces screws with lengths uniformly distributed between 4 cm and 6 cm

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
a, b = 4, 6
# Define uniform distribution parameters
uniform_dist = stats.uniform(loc=a, scale=b - a)
```

```python
# Probability of selecting a screw of exactly 5.2 cm
p_5_2 = uniform_dist.pdf(5.2)
print(f"Probability of exactly 5.2 cm screw: {p_5_2}")
```

```python
# Probability that a screw is shorter than 5 cm (CDF)
p_less_5 = uniform_dist.pdf(5)
print(f"Probability of screw being shorter than 5 cm: {p_less_5}")
```

```python
# Plot the uniform distribution
x = np.linspace(3.5, 6.5, 1000)
plt.plot(x, uniform_dist.pdf(x), label="Uniform Distribution (4 to 6)")
plt.axvline(5.2, color='r', linestyle="--", label="5.2 cm")
plt.axvline(5, color='g', linestyle="--", label="5 cm")
plt.xlabel("Screw Length (cm)")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Uniform Distribution of Screw Lengths")
plt.show()
```

```python
# Define Poisson distribution
lambda_poisson = 10  # Average sign-ups per hour
poisson_dist = stats.poisson(mu=lambda_poisson)

# Probability of exactly 15 sign-ups
p_15_signups = poisson_dist.pmf(15)
print(f"Probability of exactly 15 sign-ups: {p_15_signups:.4f}")

# Plot Poisson distribution
x = np.arange(0, 30)
plt.bar(x, poisson_dist.pmf(x), alpha=0.7, color="blue", label="Poisson PMF (λ=10)")
plt.axvline(15, color='r', linestyle="--", label="15 sign-ups")
plt.xlabel("Number of Sign-ups")
plt.ylabel("Probability Mass Function")
plt.legend()
plt.title("Poisson Distribution of Website Sign-ups")
plt.show()
```

```python
# Define binomial distribution
n_emails = 10  # Total emails
p_correct = 0.8  # Probability of correct classification
binom_dist = stats.binom(n=n_emails, p=p_correct)

# Probability of exactly 7 correct classifications
p_7_correct = binom_dist.pmf(7)
print(f"Probability of exactly 7 correct classifications: {p_7_correct:.4f}")

# Plot Binomial distribution
x = np.arange(0, 11)
plt.bar(x, binom_dist.pmf(x), alpha=0.7, color="green", label="Binomial PMF (n=10, p=0.8)")
plt.axvline(7, color='r', linestyle="--", label="7 Correct Classifications")
plt.xlabel("Number of Correctly Classified Emails")
plt.ylabel("Probability Mass Function")
plt.legend()
plt.title("Binomial Distribution of Spam Filter Performance")
plt.show()
```
