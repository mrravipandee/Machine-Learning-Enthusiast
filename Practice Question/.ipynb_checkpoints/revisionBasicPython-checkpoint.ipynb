```python
import random
# design a game 
def game():
    num = int(input('Guess the number 1 to 10: '))
    guess = random.randint(1,10)
    
    while True:
        if guess == num:
            print('Your guess was correct...')
            break
        else:
            print('Your guess was not correct...')
            num = int(input('Guess again 1 to 10: '))
            if guess == num:
                print('Guess was correct...')
                break
            else:
                print('Oops sorry try again....')


game()
```

```python
# patterns questions
def pyramid():
    n = 5
    for i in range(n+1):
        for j in range(i):
            print('*', end=' ')
        print()


def numPyranid():
    n = 5
    for i in range(n+1):
        for j in range(i):
            print(j+1, end=' ')
        print()


def primeNum():
    s = 10
    e = 100

    for i in range(s, e+1):
        for j in range(2, i):
            if i % j == 0:
                break
            else:
                print(i)

primeNum()    
```
