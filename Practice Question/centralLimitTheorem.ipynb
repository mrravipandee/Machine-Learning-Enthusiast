```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.columns
```

```python
population_data = df['SepalLengthCm']
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(population_data, bins=20, kde=True, color='blue')
plt.title('Original Distribution Sepal Lenght on Iris Datasets')
plt.xlabel('Sepal Lenght Cm')
plt.ylabel('Frequence')
plt.show()
```

### Apply Central Limit Theorem

```python
sample_means = []
sample_size = 30

for _ in range(1000):
    sample = np.random.choice(population_data, sample_size, replace=True)
    sample_means.append(np.mean(sample))
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(sample_means, bins=30, kde=True, color='purple')
plt.title('Distribution on Sample mean applied CLT')
plt.xlabel('Sample mean of Sepal Length')
plt.ylabel('Frequence')
plt.show()
```
