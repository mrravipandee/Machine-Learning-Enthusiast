```python
# 
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('imdb-top-1000.csv')
movies.head()
```

```python
# group_by() apply only those data type who are in numerical or catogrical 
genres =  movies.groupby('Genre')
```

```python
# applying agregations function on that
genres.std()
```

```python
# top 3 genres by total earning
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False) # fast & relible
```

```python
# heighest IMDM rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head()
```

```python
# find director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head()
```

```python
# find the highest rated movie of each genre
movies.groupby('Genre')['IMDB_Rating'].max()
```

```python
# each actor how many movies done
movies['Star1'].value_counts()
```

```python
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
# find total number of groups -> len
len(movies.groupby('Genre'))
```

```python
movies['Genre'].nunique()
```

```python
# check size
movies.groupby('Genre').size()
```

```python
movies['Genre'].value_counts()
```

```python
# first /last
genre = movies.groupby('Genre')
genre.first()
```

```python
genre.last()
```

```python
genre.nth(6)
```

```python
#get_group - pass those only which one come in group
genre.get_group('Horror')
```

```python
# 
genre.groups
```

```python
genre.describe()
```

```python
genre.sample()
```

```python
genre.nunique()
```

```python
# agg method - min, max, sum etc.
# genre.sum()
genre.agg({
        'Runtime': 'mean',
        'IMDB_Rating': 'mean',
        'No_of_Votes': 'sum',
        'Gross': 'sum',
        'Metascore': 'min'
})
```

```python
# agg passing list
genre.agg(['min', 'max', 'sum'])
```

```python
# looping on groups
for group,data in genre:
    print(data)
```

```python
df = pd.DataFrame(columns=movies.columns)
for group, data in genre:
    df.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])

df
```

```python
genre.apply(min)
```

```python
# movies in genre starting with, for eg. how many movie in Action genre start with A
def filter_a(group):
    # return the sum of genre start with a
    return group['Series_Title'].str.startswith('A').sum

genre.apply(filter_a)
```
