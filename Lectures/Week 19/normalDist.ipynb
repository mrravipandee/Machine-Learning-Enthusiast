```python
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
```

```python
# sample data
data = np.array([[50], [60], [70], [80], [90]])
```

```python
# apply standardrition
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
data
```

```python
scaled_data
```

```python
# plot a graph
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(7,8))
```

```python
plt.scatter(data, np.zeros_like(data), color='blue', label="Original Data")
plt.scatter(scaled_data, np.ones_like(scaled_data), color='red', label="Scaled Data")

# Labels and Title
plt.xlabel("Values")
plt.yticks([])  # Remove y-axis labels for clarity
plt.title("Effect of StandardScaler on Data")

# Show plot
plt.show()
```

```python
# Generate 1000 data points from a normal distribution
mean = 100
std_dev = 15
data = np.random.normal(mean, std_dev, 1000)
```

```python
# plot the distribution
import seaborn as sns

sns.histplot(data, kde=True, bins=30)
plt.title('Normal Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# students study prediction model
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([10, 20, 30, 40, 50])
```

```python
X.shape
```

```python
Y.shape
```

```python
# train model
model = LinearRegression()
model.fit(X, Y)
```

```python
Y_pred = model.predict
```

```python
# Plot
plt.scatter(X, Y, color='blue', label='Actual')
plt.plot(X, Y_pred, color='red', label='Predicted')
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.legend()
plt.show()
```
