```python
# 2d plot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman.head()
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# multiple ploting plot
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# label title

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit sharma VS Virat Kohli career comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# change the color of plot

plt.plot(batsman['index'], batsman['V Kohli'], color='yellow') # pass color
plt.plot(batsman['index'], batsman['RG Sharma'], color='#010101') # color in code

plt.title('Rohit sharma VS Virat Kohli career comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# change the line style(shape of lines or design)

plt.plot(batsman['index'], batsman['V Kohli'], color='yellow', linestyle='dashdot') # pass color
plt.plot(batsman['index'], batsman['RG Sharma'], color='#010101', linestyle='dashed') # color in code

plt.title('Rohit sharma VS Virat Kohli career comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# see the marker pont 

plt.plot(batsman['index'], batsman['V Kohli'], color='yellow', linestyle='dashdot', marker='D') # pass color
plt.plot(batsman['index'], batsman['RG Sharma'], color='#010101', linestyle='dashed', marker='+') # color in code

plt.title('Rohit sharma VS Virat Kohli career comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# give the axis name 
plt.plot(batsman['index'], batsman['V Kohli'], color='yellow', linestyle='dashdot', marker='D', label='Virat Kohli') # pass color
plt.plot(batsman['index'], batsman['RG Sharma'], color='#010101', linestyle='dashed', marker='+', label='Rohit Sharma') # color in code

plt.title('Rohit sharma VS Virat Kohli career comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc = 'best') # after the label use this line for show lable on the graph
```

### Bar chart

```python
batsman = pd.read_csv('batsman_season_record.csv')
plt.bar(batsman['batsman'],batsman['2015'])
```

```python
# multi bar chart
plt.bar(np.arange(batsman.shape[0]) - 0.2, batsman['2015'], width=0.2)
plt.bar(np.arange(batsman.shape[0]), batsman['2016'], width=0.2)
plt.bar(np.arange(batsman.shape[0]) + 0.2, batsman['2017'], width=0.2)
```

```python
plt.bar(np.arange(batsman.shape[0]) - 0.2, batsman['2015'], width=0.2)
plt.bar(np.arange(batsman.shape[0]), batsman['2016'], width=0.2)
plt.bar(np.arange(batsman.shape[0]) + 0.2, batsman['2017'], width=0.2)

plt.xticks(np.arange(batsman.shape[0]), batsman['batsman'])
```
