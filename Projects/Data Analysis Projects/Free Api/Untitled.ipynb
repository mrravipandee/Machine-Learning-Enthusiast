```python
import http.client
import json
import pandas as pd

# Establish the connection
conn = http.client.HTTPSConnection("realty-in-us.p.rapidapi.com")

# Headers for authentication
headers = {
    'x-rapidapi-key': "3091f55971mshc2d3a87adcdebc7p151749jsnec5ce068a4ed",
    'x-rapidapi-host': "realty-in-us.p.rapidapi.com"
}

# Make the request
conn.request("GET", "/finance/rates?loc=93505", headers=headers)

# Get the response
res = conn.getresponse()
data = res.read()

# Decode and process the response
try:
    # Decode the JSON response
    json_data = json.loads(data.decode("utf-8"))
    
    # Convert JSON data to DataFrame
    df = pd.json_normalize(json_data)
    
    # Save the DataFrame to a CSV file
    csv_path = "finance_rates.csv"
    df.to_csv(csv_path, index=False)
    print(f"CSV file saved successfully at {csv_path}")
except json.JSONDecodeError:
    print("Failed to decode JSON. Response data:", data.decode("utf-8"))
```

```python
import requests
import pandas as pd
import json

# API details
url = "https://seeking-alpha.p.rapidapi.com/market/get-earnings-calendar"
querystring = {"with_rating": "false", "currency": "USD"}
headers = {
    "x-rapidapi-key": "3091f55971mshc2d3a87adcdebc7p151749jsnec5ce068a4ed",
    "x-rapidapi-host": "seeking-alpha.p.rapidapi.com"
}

# Make the API request
response = requests.get(url, headers=headers, params=querystring)

# Load the JSON data
data = response.json()

# Normalize the data
df = pd.json_normalize(
    data['data'], 
    record_path=['attributes', 'top_companies'], 
    meta=['id', 'type', ['attributes', 'date'], ['attributes', 'ticker_count']],
    record_prefix='company_'
)

# Rename columns for better readability
df.rename(columns={
    'attributes.date': 'date',
    'attributes.ticker_count': 'ticker_count'
}, inplace=True)
```
