### `Q-1:` Find the nearest element in the array to a given integer.
```
Ex:-
a=23 and array - [10 17 24 31 38 45 52 59].
Nearest element is 24
```
Hint: Read about this function `argmin()`

````python
import numpy as np

array = np.array([10, 17, 24, 31, 38, 45, 52, 59])

a = 23

# Calculate the absolute difference
differences = np.abs(array - a)

# Find the index of the smallest difference
nearest_index = np.argmin(differences)

# Get the nearest element
nearest_element = array[nearest_index]

print('Nearest element is', nearest_element)
````

````python
# find the minimum index value of array
arr = np.random.randint(1, 11, 10)

min_index = np.argmin(arr)

print('Array are -',arr)
print('Minimum index is:',min_index,' value is: ',arr[min_index])
````

### `Q-4:` Your are given an array which is havig some nan value. You job is to fill those nan values with most common element in the array.
```
arr=np.array([[1,2,np.nan],[4,2,6],[np.nan,np.nan,5]])

```

````python
# code here
arr = np.array([[1,2,np.nan],[4,2,6],[np.nan,np.nan,5]])

non_nan_values =  arr[~np.isnan(arr)] # give 1d

unique, counts = np.unique(non_nan_values, return_counts=True)
most_common = unique[np.argmax(counts)]

arr[np.isnan(arr)] = most_common

print(arr)

````

### `Q-7:` Plot this below equation.

$$\large y = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

**Note: This equation is called tanh activation function. In deep learning, many times this function is used. If you find some difference between the sigmoid function and this tanh function, note that to your notebook.**

````python
# code here

import numpy as np
import matplotlib.pyplot as plt

# Define the Sigmoid function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Define the Tanh function and its derivative
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# Generate a range of values for x
x = np.linspace(-5, 5, 500)

# Compute Sigmoid, Tanh, and their derivatives
sigmoid_values = sigmoid(x)
sigmoid_derivative_values = sigmoid_derivative(x)

tanh_values = tanh(x)
tanh_derivative_values = tanh_derivative(x)

# Plot the activation functions and their derivatives
plt.figure(figsize=(12, 8))

# Sigmoid plot
plt.subplot(2, 2, 1)
plt.plot(x, sigmoid_values, label="Sigmoid")
plt.title("Sigmoid Activation Function")
plt.grid()
plt.legend()

# Sigmoid derivative plot
plt.subplot(2, 2, 2)
plt.plot(x, sigmoid_derivative_values, label="Sigmoid Derivative", color='orange')
plt.title("Sigmoid Derivative")
plt.grid()
plt.legend()

# Tanh plot
plt.subplot(2, 2, 3)
plt.plot(x, tanh_values, label="Tanh", color='green')
plt.title("Tanh Activation Function")
plt.grid()
plt.legend()

# Tanh derivative plot
plt.subplot(2, 2, 4)
plt.plot(x, tanh_derivative_values, label="Tanh Derivative", color='red')
plt.title("Tanh Derivative")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
````

### `Q-9:` Write a program implement Boradcasting Rule to check if two array can be added or not.
Given tuples of shapes.
```
shape of a- (3,2,2)
shape of b- (2,2)

check_broadcast(a, b) -> return Boolean (True if can broadcasted, False other wise.)
```

````python
# code here

def check_boradcasting(shape_a, shape_b):
    reverse_a = shape_a[::-1]
    reverse_b = shape_b[::-1]

    # compare dimanetion from the end
    for i in range(max(len(reverse_a), len(reverse_b))):
        dim_a = reverse_a[i] if i < len(reverse_a) else 1
        dim_b = reverse_b[i] if i < len(reverse_b) else 1

    # check boradcast rule
    if dim_a != dim_b and dim_a != 1 and dim_b != 1:
        return False
    return True


shape_a = (3,2,2)
shape_b = (2,2)

check_boradcasting(shape_a, shape_b)
````
